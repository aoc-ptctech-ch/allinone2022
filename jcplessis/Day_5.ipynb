```python
def parse_stacks(input):
  lines = input.split("\n")
  result = [[] for x in range(int(len(lines[0])/3))]
  for line in lines:
    if len(line) == 0:
      return result
    stack = 0
    while len(line) > 0:
      line, chars = line[4:], line[:4]
      char = chars[1]
      if (char != " " and char >= "A" and char <= "Z"):
        result[stack].insert(0, char)
      stack += 1
    
  return result

def move_crate(stacks, from_column, to_column):
  stacks[to_column-1].append(stacks[from_column-1].pop())
  return stacks

def parse_move(stacks, move):
  iterations = int(move.split(" ")[1])
  from_column = int(move.split(" ")[3])
  to_column = int(move.split(" ")[5])
  for i in range(iterations):
    stacks = move_crate(stacks, from_column, to_column)
  return stacks

def parse_move_CrateMover9001(stacks, move):
  number = int(move.split(" ")[1])
  from_column = int(move.split(" ")[3]) - 1
  to_column = int(move.split(" ")[5]) - 1
  stacks[from_column], removed = stacks[from_column][:-1*number], stacks[from_column][-1*number:]
  stacks[to_column].extend(removed)
  return stacks

    
def list_top_crates(stacks):
  result = ""
  for stack in stacks:
    if len(stack) > 0:
      result += stack[-1]
    else:
      result += " "
  return result

def apply_all_moves(input, CrateMover9001=False):
  stacks = parse_stacks(input)
  moves = input.split("\n\n")[1]
  for move in moves.split("\n"):
    if CrateMover9001:
      stacks = parse_move_CrateMover9001(stacks, move)
    else:
      stacks = parse_move(stacks, move)
  return list_top_crates(stacks)

```

```python
import unittest

example = """    [D]    
[N] [C]    
[Z] [M] [P]
 1   2   3 

move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2"""

class TestAcceptance(unittest.TestCase):

    def test_parse_stack(self):
      self.assertEqual([["Z", "N"], ["M", "C", "D"], ["P"]], parse_stacks(example))

    def test_move_stack(self):
      self.assertEqual([[], ["A"]], move_crate([["A"], []], 1, 2))

    def test_parse_stack_move(self):
      self.assertEqual([[], ["B", "A"]], parse_move([["A", "B"], []], "move 2 from 1 to 2"))

    def test_list_top_crates(self):
      self.assertEqual("BC", list_top_crates([["A", "B"], ["C"]]))
      self.assertEqual("B C", list_top_crates([["A", "B"], [], ["C"]]))


    def test_apply_all_moves(self):
      self.assertEqual("CMZ", apply_all_moves(example))

    def test_parse_stack_move_multiple(self):
      self.assertEqual([[], ["A", "B"]], parse_move_CrateMover9001([["A", "B"], []], "move 2 from 1 to 2"))

    def test_apply_all_moves_withCrateMover9001(self):
      self.assertEqual("MCD", apply_all_moves(example, CrateMover9001=True))



unittest.main(argv=[''], verbosity=2, exit=False)
```
