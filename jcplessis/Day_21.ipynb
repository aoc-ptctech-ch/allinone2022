```python
from pprint import pprint

def return_value_func(val):
  return lambda: int(val)

def return_eval_func(first, op, second):
  return lambda: eval(f"{first}() {op} {second}()")

def solve(input):
  for line in input.split("\n"):
    name, value = line.split(": ")
    if value.isdigit():
      globals()[name] = return_value_func(value)
    else:
      first, op, second = value.split(" ")
      globals()[name] = return_eval_func(first, op, second)
  
  return globals()["root"]()



def solve_equality(input):
  value = 1
  result = compute(input, value)
  while result > 0:
    value *= 2 
    result = compute(input, value)
    print(f"{value} -> {result}")
  min, max = value/2, value
  while True:
    print(f"Should be between {min} and {max}")
    med = min + (max-min)/2
    result = compute(input, med)
    print(f"{med} - {result}")
    if result > 0:
      min = med
    if result < 0:
      max = med
    if result == 0:
      return med


def compute(input, my_input):
  for line in input.split("\n"):
    name, value = line.split(": ")
    if value.isdigit():
      if name == "humn":
        value = my_input
      globals()[name] = return_value_func(value)
    else:
      first, op, second = value.split(" ")
      if name == "root":
        op = " - "
      globals()[name] = return_eval_func(first, op, second)
  
  return globals()["root"]()
```
