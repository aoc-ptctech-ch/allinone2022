```python
def score(opponent, me):
    win_score = {
        "AX" : 3,
        "AY" : 6,
        "AZ" : 0,
        "BX" : 0,
        "BY" : 3,
        "BZ" : 6,
        "CX" : 6,
        "CY" : 0,
        "CZ" : 3
    }[opponent+me]
    choice_score = {
        "X":1,
        "Y":2,
        "Z":3
    }[me]

    return win_score + choice_score

def score_all(matches):
    res = 0
    for match in matches.split("\n"):
        res += score(*match.split(" "))
    return res

def score2(opponent, result):
    win_score = {
        "X" : 0,
        "Y" : 3,
        "Z" : 6
    }[result]
 
    my_choice = {
            "AX" : "C",
            "AY" : "A",
            "AZ" : "B",
            "BX" : "A",
            "BY" : "B",
            "BZ" : "C",
            "CX" : "B",
            "CY" : "C",
            "CZ" : "A"
        }[opponent+result]
 
    choice_score = {
        "A":1,
        "B":2,
        "C":3
    }[my_choice]
 
    return win_score + choice_score
 
def score_all2(matches):
    res = 0
    for match in matches.split("\n"):
        res += score2(*match.split(" "))
    return res
```

```python
import unittest
 
class TestStringMethods(unittest.TestCase):
 
    def test_one_match(self):
        self.assertEqual(3+1, score("A", "X"))
        self.assertEqual(6+2, score("A", "Y"))
 
    def test_all_match(self):
            self.assertEqual(15, score_all("""A Y
B X
C Z"""))
 
    def test_one_match2(self):
        self.assertEqual(3+1, score2("A", "Y"))
 
 
    def test_all_match2(self):
            self.assertEqual(12, score_all2("""A Y
B X
C Z"""))
 
unittest.main(argv=[''], verbosity=2, exit=False)
```
