```python
from functools import cmp_to_key

def compare(left, right):
  return test_order(eval(left), eval(right))

def test_order(left, right):
  left = left[:]
  right = right[:]
  if len(left) == 0 and len(right) == 0:
    return 0
  if len(left) == 0 and len(right) > 0:
    return -1
  if len(left) > 0 and len(right) == 0:
    return 1

  l1 = left.pop(0)
  r1 = right.pop(0)
  if isinstance(l1, list) and isinstance(r1, int):
    left.insert(0, l1)
    right.insert(0, [r1])
    return test_order(left, right)
  if isinstance(l1, int) and isinstance(r1, list):
    left.insert(0, [l1])
    right.insert(0, r1)
    return test_order(left, right)

  if isinstance(l1, list) and isinstance(r1, list):
    sub_result = test_order(l1, r1)
    if sub_result == 0:
      return test_order(left, right)
    else:
      return sub_result

  if isinstance(l1, int) and isinstance(r1, int):
    if l1 < r1:
      return -1
    if l1 == r1:
      return test_order(left, right)
    return 1


def sum_right_order_indices(input):
  result = 0
  pairs = input.split("\n\n")
  for index, pair in enumerate(pairs):
    if compare(*pair.split("\n")) == -1:
      result += index+1
  return result

def sorted_packets_with_dividers(input):
  to_sort = [
      [[2]],
      [[6]]
  ]
  to_sort.extend([eval(line) for line in input.replace("\n\n", "\n").split("\n")])
  to_sort.sort(key=cmp_to_key(test_order))
  return to_sort

def decoder_key(input):
  packets = sorted_packets_with_dividers(input)
  return (packets.index([[2]]) + 1) * (packets.index([[6]]) + 1)
```

```python
import unittest

example = """[1,1,3,1,1]
[1,1,5,1,1]

[[1],[2,3,4]]
[[1],4]

[9]
[[8,7,6]]

[[4,4],4,4]
[[4,4],4,4,4]

[7,7,7,7]
[7,7,7]

[]
[3]

[[[]]]
[[]]

[1,[2,[3,[4,[5,6,7]]]],8,9]
[1,[2,[3,[4,[5,6,0]]]],8,9]"""

class TestAcceptance(unittest.TestCase):

    def test_is_in_right_order(self):
      self.assertEqual(-1, compare("[1,1,3,1,1]", "[1,1,5,1,1]"))
      self.assertEqual(-1, compare("[[1],[2,3,4]]", "[[1],4]"))
      self.assertEqual(1, compare("[9]", "[[8,7,6]]"))

    def test_sum_right_order_indices(self):
      self.assertEqual(13, sum_right_order_indices(example))

    def test_sort(self):
      self.assertEqual([
          [],
          [[]],
          [[[]]],
          [1,1,3,1,1],
          [1,1,5,1,1],
          [[1],[2,3,4]],
          [1,[2,[3,[4,[5,6,0]]]],8,9],
          [1,[2,[3,[4,[5,6,7]]]],8,9],
          [[1],4],
          [[2]],
          [3],
          [[4,4],4,4],
          [[4,4],4,4,4],
          [[6]],
          [7,7,7],
          [7,7,7,7],
          [[8,7,6]],
          [9]
      ], sorted_packets_with_dividers(example))

    def test_decoder_key(self):
      self.assertEqual(140, decoder_key(example))

unittest.main(argv=[''], verbosity=2, exit=False)
```
