```python
def find_duplicate(rucksack):
  limit = int(len(rucksack)/2)
  first_half = rucksack[:limit]
  second_half = rucksack[limit:]
  return [x for x in first_half if x in second_half][0]

def priority(character):
  result = ord(character) - ord("a") + 1
  if(result < 0):
    result = ord(character) - ord("A") + 1 + priority("z")
  return result

def priorities_of_doubles(rucksacks):
  result = 0
  for rucksack in rucksacks.split("\n"):
    result += priority(
        find_duplicate(rucksack)
    )
  return result

def find_badge(rucksack1, rucksack2, rucksack3):
   return [x for x in rucksack1 if x in rucksack2 and x in rucksack3][0]

def find_badge_sums(rucksacks):
  result = 0
  rucksacks = rucksacks.split("\n")
  while len(rucksacks) > 0:
    group, rucksacks = rucksacks[:3], rucksacks[3:]
    result += priority(
        find_badge(*group)
    )
  return result
```
